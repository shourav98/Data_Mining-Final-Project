```python
gluc_dict= dict(zip(df['gluc'].unique(),range(len(df['gluc'].unique()))))
gluc_dict
```

```python
df['gluc_nw'] = df['gluc'].apply(lambda x: gluc_dict[x])
df
```

```python
weight_dict= dict(zip(df['weight'].unique(),range(len(df['weight'].unique()))))
weight_dict
```

```python
df['weight_nw'] = df['weight'].apply(lambda x: weight_dict[x])
df
```

```python
ap_hi_dict= dict(zip(df['ap_hi'].unique(),range(len(df['ap_hi'].unique()))))
ap_hi_dict
```

```python
df['ap_hi_nw'] = df['ap_hi'].apply(lambda x: ap_hi_dict[x])
df
```

```python
ap_lo_dict= dict(zip(df['ap_lo'].unique(),range(len(df['ap_lo'].unique()))))
ap_lo_dict
```

```python
df['ap_lo_nw'] = df['ap_lo'].apply(lambda x: ap_lo_dict[x])
df
```

```python
cholesterol_dict= dict(zip(df['cholesterol'].unique(),range(len(df['cholesterol'].unique()))))
cholesterol_dict
```

```python
df['cholesterol_nw'] = df['cholesterol'].apply(lambda x: cholesterol_dict[x])
df
```

```python
cardio_dict= dict(zip(df['cardio'].unique(),range(len(df['cardio'].unique()))))
cardio_dict
```

```python
df['cardio_nw'] = df['cardio'].apply(lambda x: cardio_dict[x])
df
```

```python
label=list(df['cardio_nw'])
label
```

```python
def return_weight_gluc_ap_hi_ap_lo_cholesterol_tuple(gluc_nw,weight_nw,ap_hi_nw,ap_lo_nw,cholesterol_nw):
    return gluc_nw,weight_nw,ap_hi_nw,ap_lo_nw,cholesterol_nw
```

```python
features=[]
for i in range (len(df['weight_nw'])):
    features.append(return_weight_gluc_ap_hi_ap_lo_cholesterol_tuple(df['gluc_nw'][i],df['weight_nw'][i],df['ap_hi_nw'][i],df['ap_lo_nw'][i],df['cholesterol_nw'][i]))
features
```

```python
Model=GaussianNB()
```

```python
Model.fit(features,label)
```

```python
print("Enter the value of Glucose_level,Weight,Sistolic_pressure,diostolic_pressure,Cholesterol_level ")
```

```python
predicted= Model.predict([[1,70,90,140,3]])
```
